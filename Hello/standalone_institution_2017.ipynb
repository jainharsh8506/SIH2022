```python
df_stand_accr_2017_sort_si_2017=df_stand_accr_2017_si_2017.sort_values(by='percentage',ascending=False)
```

```python
def Accreditation_2017():
    df_accreditation_si_2017=df_stand_accr_2017_sort_si_2017.rename(columns={'aishe_code':'AISHE Code', 'name':'Standalone Institution Name', 'accreditation_body':'Accreditation Body', 'score':'Score', 'max_score': 'Maximum Score', 'percentage':'Percentage'})
    return df_accreditation_si_2017[['AISHE Code','Standalone Institution Name','Accreditation Body','Score','Maximum Score','Percentage']].head(10)
Accreditation_2017().to_csv("standalone_accreditation_2017")
```
