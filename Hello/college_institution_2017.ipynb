```python
import pandas as pd
```

```python
df_accr_c_2017=pd.read_csv("static/csv/2017/accreditation.csv")
df_accr_c_2017
```

```python
df_clg_accr_c_2017=pd.read_csv("static/csv/2017/college_institution_accreditation.csv")
df_clg_accr_c_2017
```

```python
df_clg_accr_accr_c_2017=pd.merge(df_accr_c_2017,df_clg_accr_c_2017,left_on='id',right_on='accreditation_id',how='inner').drop(['id'],axis=1)
df_clg_accr_accr_c_2017
```

```python
df_clg_c_2017=pd.read_csv("static/csv/2017/college_institution.csv")
df_clg_c_2017
```

```python
df_clg_accr_merge_c_2017=pd.merge(df_clg_accr_accr_c_2017,df_clg_c_2017,left_on='college_institution_id',right_on='id',how='inner').drop(['id'],axis=1)
df_clg_accr_merge_c_2017
```

```python
df_clg_accr_merge_c_2017['percentage']=(df_clg_accr_merge_c_2017['score'] / df_clg_accr_merge_c_2017['max_score'])*100
df_clg_accr_merge_c_2017
```

```python
df_clg_accr_merge_c_2017['score'].fillna(1,inplace= True)
```

```python
df_clg_accr_merge_c_2017['max_score'].fillna(1,inplace= True)
```

```python
df_clg_accr_merge_c_2017['percentage'].fillna(1,inplace= True)
```

```python
df_clg_accr_merge_c_2017
```

```python
df_clg_accr_c_2017=df_clg_accr_merge_c_2017[['college_institution_id','name','aishe_code','accreditation_body','survey_year_x','has_score','score','max_score','percentage','infrastructure_id']]
df_clg_accr_c_2017
```

```python
df_clg_accr_c_2017_sort=df_clg_accr_c_2017.sort_values(by='percentage',ascending=False)
df_clg_accr_c_2017_sort
```

```python
df_infr_c_2017=pd.read_csv("static/csv/2017/infrastructure.csv")
df_infr_c_2017
```

```python
df_clg_accr_infr_c_2017=pd.merge(df_clg_accr_c_2017_sort,df_infr_c_2017,left_on='infrastructure_id',right_on='id',how='inner').drop(['id'],axis=1)
df_clg_accr_infr_c_2017
```

```python
df_clg_accr_infrastructure_c_2017=df_clg_accr_infr_c_2017[['college_institution_id','name','aishe_code','accreditation_body','survey_year_x','has_score',
                                             'score','max_score','percentage','infrastructure_id','playground','library','connectivity_nkn',
                                             'laboratory','indoor_stadium','cafeteria','computer_center','campus_friendly']]
```

```python
df_clg_accr_infrastructure_c_2017.playground = df_clg_accr_infrastructure_c_2017.playground.replace({True:1,False:0})
```

```python
df_clg_accr_infrastructure_c_2017.library = df_clg_accr_infrastructure_c_2017.library.replace({True:1,False:0})
```

```python
df_clg_accr_infrastructure_c_2017.connectivity_nkn = df_clg_accr_infrastructure_c_2017.connectivity_nkn.replace({True:1,False:0})
```

```python
df_clg_accr_infrastructure_c_2017.laboratory = df_clg_accr_infrastructure_c_2017.laboratory.replace({True:1,False:0})
```

```python
df_clg_accr_infrastructure_c_2017.indoor_stadium = df_clg_accr_infrastructure_c_2017.indoor_stadium.replace({True:1,False:0})
```

```python
df_clg_accr_infrastructure_c_2017.cafeteria = df_clg_accr_infrastructure_c_2017.cafeteria.replace({True:1,False:0})
```

```python
df_clg_accr_infrastructure_c_2017.computer_center = df_clg_accr_infrastructure_c_2017.computer_center.replace({True:1,False:0})
```

```python
df_clg_accr_infrastructure_c_2017.campus_friendly = df_clg_accr_infrastructure_c_2017.campus_friendly.replace({True:1,False:0})
```

```python
df_clg_accr_infrastructure_c_2017
```

```python
df_filtered_infra_c_2017 = df_clg_accr_infrastructure_c_2017[['college_institution_id','infrastructure_id','playground','library','connectivity_nkn',	'laboratory','indoor_stadium','cafeteria','computer_center',	'campus_friendly']]
df_filtered_infra_c_2017
```

```python
df_filtered_infra_c_2017['infra_count']=df_filtered_infra_c_2017["playground"] + df_filtered_infra_c_2017["library"]+ df_filtered_infra_c_2017["connectivity_nkn"] + df_filtered_infra_c_2017["indoor_stadium"] + df_filtered_infra_c_2017["laboratory"] + df_filtered_infra_c_2017["cafeteria"] + df_filtered_infra_c_2017["computer_center"] + df_filtered_infra_c_2017["campus_friendly"]
df_filtered_infra_c_2017
```

```python
find_infra=['playground','library','laboratory','indoor_stadium','connectivity_nkn','cafeteria','computer_center','campus_friendly']
```

```python
for i in range(len(find_infra)):
    print(find_infra[i])
```

```python
df_filtered_infra_c_2017[find_infra]
```

```python
def dynamic_infra_sum(df_filtered_infra_c_2017,find_infra):
    return df_filtered_infra_c_2017[find_infra].sum(axis=1)
```

```python
df_clg_accr_infrastructure_c_2017['infra_count']=dynamic_infra_sum(df_clg_accr_infrastructure_c_2017,find_infra)
df_clg_accr_infrastructure_c_2017
```

```python
df_clg_accr_infrastructure_sort_c_2017=df_clg_accr_infrastructure_c_2017.sort_values(['percentage','infra_count'],ascending=False)
df_clg_accr_infrastructure_sort_c_2017
```

```python
df_examination_result_c_2017 = pd.read_csv("static/csv/2017/examination_result.csv")
df_examination_result_c_2017
```

```python
df_edu_inst_c_2017=pd.read_csv("static/csv/2017/educational_institution_course.csv")
df_edu_inst_c_2017
```

```python
df_examination_result_merge_c_2017 =pd.merge(df_edu_inst_c_2017,df_examination_result_c_2017,left_on='course_id',right_on='course_id',how='inner')
df_examination_result_merge_c_2017
```

```python
df_examination_result_merge_filt_c_2017=df_examination_result_merge_c_2017.drop(['id','appeared_female','passed_female','first_class_passed_total','first_class_passed_female','broad_discipline_group_id'],axis= 1)
df_examination_result_merge_filt_c_2017
```

```python
df_examination_result_merge_filt_c_2017['pass_percentage']=(df_examination_result_merge_filt_c_2017['passed_total'] / df_examination_result_merge_filt_c_2017['appeared_total'])*100
df_examination_result_merge_filt_c_2017
```

```python
df_clg_accr_infra_ex_c_2017=pd.merge(df_examination_result_merge_filt_c_2017,df_clg_accr_infrastructure_sort_c_2017,left_on='institution_id',right_on= 'college_institution_id').drop(['institution_id'],axis=1)
df_clg_accr_infra_ex_c_2017
```

```python
df_clg_accr_infra_exam_c_2017=df_clg_accr_infra_ex_c_2017.sort_values(['percentage','infra_count','pass_percentage'],ascending=False)
df_clg_accr_infra_exam_c_2017
```

```python
df_clg_accr_infra_exam_sort_c_2017=df_clg_accr_infra_exam_c_2017[[ 'appeared_total', 'passed_total','aishe_code','college_institution_id','name', 'state_code','department_id','discipline', 'faculty_id',  'course_id', 'survey_year','examination_result_id', 'percentage','infra_count','pass_percentage']]
```

```python
def result_2017():
    df_exam_result_c_2017=df_clg_accr_infra_exam_sort_c_2017.rename(columns={'aishe_code':'AISHE Code', 'name':'College Institution Name','discipline':'Discipline', 'appeared_total':'Appeared Students', 'passed_total':'Passed Student', 'pass_percentage': 'Passing Percentage'})
    return df_exam_result_c_2017[['AISHE Code','College Institution Name','Discipline','Appeared Students','Passed Student','Passing Percentage']].head(10)
result_2017().to_csv("college_institution_result_2017")
```
