```python
df_accr_si_2018=pd.read_csv("static/csv/2018/accreditation.csv")
```

```python
df_stand_accr_accr_si_2018=pd.merge(df_accr_si_2018,df_stand_accr_si_2018,left_on='id',right_on='accreditation_id',how='inner').drop(['id'],axis=1)
```

```python
df_stand_accr_merge_si_2018
```

```python
df_stand_accr_merge_si_2018['score'].fillna(1,inplace= True)
```

```python
df_stand_accr_2018_si_2018=df_stand_accr_merge_si_2018[['standalone_institution_id','name','aishe_code','accreditation_body','survey_year','has_score','score','max_score','percentage','infrastructure_id']]
```

```python
df_stand_accr_2018_sort_si_2018=df_stand_accr_2018_si_2018.sort_values(by='percentage',ascending=False)
```

```python
def Accreditation_2018():
    df_accreditation_si_2018=df_stand_accr_2018_sort_si_2018.rename(columns={'aishe_code':'AISHE Code', 'name':'Standalone Institution Name', 'accreditation_body':'Accreditation Body', 'score':'Score', 'max_score': 'Maximum Score', 'percentage':'Percentage'})
    return df_accreditation_si_2018[['AISHE Code','Standalone Institution Name','Accreditation Body','Score','Maximum Score','Percentage']].head(10)
```

```python
Accreditation_2018().to_csv("standalone_accreditation_2018")
```

```python
df_infr_si_2018=pd.read_csv("static/csv/2018/infrastructure.csv")
```

```python
df_stand_accr_infrastructure_si_2018=df_stand_accr_infr_si_2018[['standalone_institution_id','name','aishe_code','accreditation_body','survey_year','has_score',
                                             'score','max_score','percentage','infrastructure_id','playground','library','connectivity_nkn',
                                             'laboratory','indoor_stadium','cafeteria','computer_center','campus_friendly']]
```

```python
df_stand_accr_infrastructure_si_2018.playground = df_stand_accr_infrastructure_si_2018.playground.replace({True:1,False:0})
df_stand_accr_infrastructure_si_2018.library = df_stand_accr_infrastructure_si_2018.playground.replace({True:1,False:0})
df_stand_accr_infrastructure_si_2018.connectivity_nkn = df_stand_accr_infrastructure_si_2018.connectivity_nkn.replace({True:1,False:0})
df_stand_accr_infrastructure_si_2018.laboratory = df_stand_accr_infrastructure_si_2018.laboratory.replace({True:1,False:0})
df_stand_accr_infrastructure_si_2018.indoor_stadium = df_stand_accr_infrastructure_si_2018.indoor_stadium.replace({True:1,False:0})
df_stand_accr_infrastructure_si_2018.cafeteria = df_stand_accr_infrastructure_si_2018.cafeteria.replace({True:1,False:0})
df_stand_accr_infrastructure_si_2018.computer_center = df_stand_accr_infrastructure_si_2018.computer_center.replace({True:1,False:0})
df_stand_accr_infrastructure_si_2018.campus_friendly = df_stand_accr_infrastructure_si_2018.campus_friendly.replace({True:1,False:0})
```
