```python
import pandas as pd
```

```python
df_edu_ins_course = pd.read_csv("static/csv/2019/educational_institution_course.csv")
df_edu_ins_course
```

```python
df_exam_res = pd.read_csv("static/csv/2019/examination_result.csv")
df_exam_res
```

```python
df_edu_ins_exam_res = pd.merge(df_edu_ins_course, df_exam_res,left_on= 'course_id', right_on='course_id',how= 'inner')
df_edu_ins_exam_res
```

```python
df_edu_ins_exam_res.columns
```

```python
df_edu_ins_exam_res1 = df_edu_ins_exam_res.drop(['id','appeared_female','passed_female','first_class_passed_total','first_class_passed_female','broad_discipline_group_id'],axis= 1)
df_edu_ins_exam_res1
```

```python
df_edu_ins_exam_res1['pass percentage']=(df_edu_ins_exam_res1['passed_total'] / df_edu_ins_exam_res1['appeared_total'])*100
df_edu_ins_exam_res1
```

```python
df_course = pd.read_csv("static/csv/2019/course.csv")
df_course
```

```python
df_course1 = df_course.drop(['mode_id','only_through_colleges','level_id','programme_id','discipline','type_id','approving_statutory_body_id','approving_university_id','broad_discipline_group_category_id','admission_criterion_id'],axis=1)
df_course1
```

```python
df_ins_res_course = pd.merge(df_edu_ins_exam_res1, df_course1, left_on = 'course_id', right_on = 'id', how = 'inner').drop(['id'],axis=1)
df_ins_res_course
```

```python
df_ins_res_course.columns
```

```python
df_ins_res_course.drop(['course_mode_id','course_level_id','programme_id'],axis= 1)
```

```python
df_clg_ins= pd.read_csv("static/csv/2019/college_institution.csv")
df_clg_ins
```

```python
df_clg_ins1 = df_clg_ins.drop(['state_code','survey_year'],axis=1)
```

```python
df_ins_res_cou_clg = pd.merge(df_ins_res_course, df_clg_ins1, left_on ='institution_id', right_on = 'id', how = 'inner').drop(['id'],axis=1)
df_ins_res_cou_clg.columns
```

```python
df_ins_clg_sort=df_ins_res_cou_clg.sort_values(by='pass percentage',ascending=False)
df_ins_clg_sort
```

```python
df_infra1=df_infra[['id','playground','library','connectivity_nkn','laboratory','indoor_stadium','cafeteria','computer_center','campus_friendly']]
```
