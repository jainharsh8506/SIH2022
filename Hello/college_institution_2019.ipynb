```python
find_infra=['playground','library','laboratory','indoor_stadium','connectivity_nkn','cafeteria','computer_center','campus_friendly']
```

```python
def dynamic_infra_sum(df_filtered_infra_c_2019,find_infra):
    return df_filtered_infra_c_2019[find_infra].sum(axis=1)
```

```python
df_clg_accr_infrastructure_c_2019['infra_count']=dynamic_infra_sum(df_clg_accr_infrastructure_c_2019,find_infra)
```
