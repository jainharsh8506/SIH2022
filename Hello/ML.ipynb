```python
frames=[df_top_accreditation_2017,df_top_accreditation_2018,df_top_accreditation_2019]
```

```python
df_result=pd.concat(frames)
df_result

```

```python
df_result.rename(columns={"survey_year_x" : "Survey_Year","percentage":"aggregate_score", "name":"College_name"}, inplace = True)
```

```python
df_result_final=df_result.drop(['Unnamed: 0','infrastructure_id','has_score','score','max_score','accreditation_body'],axis=1)
```

```python
df_1= df_result_final.query("College_name == 'Tata Institute of Social Sciences, Mumbai'")
```

```python
df_1
```

```python
x=df_1.iloc[:, 3:4].values
y=df_1.iloc[: , 4].values
```

```python
print(x)
```

```python
print(y)
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lin = LinearRegression()
```

```python
lin.fit(x,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly= PolynomialFeatures(degree=4)
```

```python
x_poly=poly.fit_transform(x)
```

```python
poly.fit(x_poly, y)
lin2 = LinearRegression()
lin2.fit(x_poly, y)
```

```python
plt.scatter(x, y, color = 'blue')
  
plt.plot(x, lin.predict(x), color = 'red')
plt.title('Linear Regression')
plt.xlabel('YEAR')
plt.ylabel('SCORE')
  
plt.show()
```

```python
plt.scatter(x,y,color='red')
```

```python
plt.plot(x, lin2.predict(poly.fit_transform(x)), color = 'red')
plt.title('Polynomial Regression')
plt.xlabel('Survey_Year')
plt.ylabel('Percentage')
plt.show()
```
