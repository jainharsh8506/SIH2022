```python
import pandas as pd
```

```python
df_accr_si_2019=pd.read_csv("static/csv/2019/accreditation.csv")
```

```python
df_accr_si_2019
```

```python
df_stand_accr_si_2019=pd.read_csv("static/csv/2019/standalone_institution_accreditation.csv")
```

```python
df_stand_accr_accr_si_2019=pd.merge(df_accr_si_2019,df_stand_accr_si_2019,left_on='id',right_on='accreditation_id',how='inner').drop(['id'],axis=1)
```

```python
df_stand_accr_accr_si_2019
```

```python
df_stand_si_2019=pd.read_csv("static/csv/2019/standalone_institution.csv")
```

```python
df_stand_accr_merge_si_2019=pd.merge(df_stand_accr_accr_si_2019,df_stand_si_2019,left_on='standalone_institution_id',right_on='id',how='inner').drop(['id'],axis=1)
```

```python
df_stand_accr_merge_si_2019
```

```python
df_stand_accr_merge_si_2019=df_stand_accr_merge_si_2019.rename(columns={'survey_year_x':'survey_year'})
```

```python
df_stand_accr_merge_si_2019['percentage']=(df_stand_accr_merge_si_2019['score'] / df_stand_accr_merge_si_2019['max_score'])*100
```

```python
df_stand_accr_merge_si_2019
```

```python
df_stand_accr_merge_si_2019['score'].fillna(1,inplace= True)
```

```python
df_stand_accr_merge_si_2019['max_score'].fillna(1,inplace= True)
```

```python
df_stand_accr_merge_si_2019['percentage'].fillna(1,inplace= True)
```

```python
df_stand_accr_2019_si_2019=df_stand_accr_merge_si_2019[['standalone_institution_id','name','aishe_code','accreditation_body','survey_year','has_score','score','max_score','percentage','infrastructure_id']]
```

```python
df_stand_accr_2019_sort_si_2019=df_stand_accr_2019_si_2019.sort_values(by='percentage',ascending=False)
```

```python
df_infr_si_2019=pd.read_csv("static/csv/2019/infrastructure.csv")
```

```python
df_stand_accr_infr_si_2019=pd.merge(df_stand_accr_2019_sort_si_2019,df_infr_si_2019,left_on='infrastructure_id',right_on='id',how='inner').drop(['id'],axis=1)
```

```python
df_stand_accr_infrastructure_si_2019=df_stand_accr_infr_si_2019[['standalone_institution_id','name','aishe_code','accreditation_body','survey_year','has_score',
                                             'score','max_score','percentage','infrastructure_id','playground','library','connectivity_nkn',
                                             'laboratory','indoor_stadium','cafeteria','computer_center','campus_friendly']]
```

```python
df_stand_accr_infrastructure_si_2019.playground = df_stand_accr_infrastructure_si_2019.playground.replace({True:1,False:0})
df_stand_accr_infrastructure_si_2019.library = df_stand_accr_infrastructure_si_2019.playground.replace({True:1,False:0})
df_stand_accr_infrastructure_si_2019.connectivity_nkn = df_stand_accr_infrastructure_si_2019.connectivity_nkn.replace({True:1,False:0})
df_stand_accr_infrastructure_si_2019.laboratory = df_stand_accr_infrastructure_si_2019.laboratory.replace({True:1,False:0})
df_stand_accr_infrastructure_si_2019.indoor_stadium = df_stand_accr_infrastructure_si_2019.indoor_stadium.replace({True:1,False:0})
df_stand_accr_infrastructure_si_2019.cafeteria = df_stand_accr_infrastructure_si_2019.cafeteria.replace({True:1,False:0})
df_stand_accr_infrastructure_si_2019.computer_center = df_stand_accr_infrastructure_si_2019.computer_center.replace({True:1,False:0})
df_stand_accr_infrastructure_si_2019.campus_friendly = df_stand_accr_infrastructure_si_2019.campus_friendly.replace({True:1,False:0})
```

```python
df_stand_accr_infrastructure_si_2019
```

```python
df_filtered_infra_si_2019= df_stand_accr_infrastructure_si_2019[['standalone_institution_id','infrastructure_id','playground','library','connectivity_nkn',	'laboratory','indoor_stadium','cafeteria','computer_center',	'campus_friendly']]
```

```python
df_filtered_infra_si_2019
```

```python
df_filtered_infra_si_2019['infra_count']=df_filtered_infra_si_2019.playground + df_filtered_infra_si_2019.library + df_filtered_infra_si_2019.connectivity_nkn + df_filtered_infra_si_2019.indoor_stadium + df_filtered_infra_si_2019.laboratory + df_filtered_infra_si_2019.cafeteria + df_filtered_infra_si_2019.computer_center + df_filtered_infra_si_2019.campus_friendly
```

```python
df_filtered_infra_si_2019
```

```python
find_infra_si_2019=['playground','library','laboratory','indoor_stadium','connectivity_nkn','cafeteria','computer_center','campus_friendly']
```

```python
for i in range(len(find_infra_si_2019)):
    print(find_infra_si_2019[i])
```

```python
df_filtered_infra_si_2019[find_infra_si_2019]
```

```python
df_filtered_infra_si_2019[find_infra_si_2019].sum(axis=1)
```

```python
def dynamic_infra_sum(df_filtered_infra_si_2019,find_infra_si_2019):
    return df_filtered_infra_si_2019[find_infra_si_2019].sum(axis=1)
```

```python
df_stand_accr_infrastructure_si_2019['infra_count']=dynamic_infra_sum(df_stand_accr_infrastructure_si_2019,find_infra_si_2019)
```

```python
df_stand_accr_infrastructure_si_2019
```

```python
df_stand_accr_infrastructure_si_2019[['standalone_institution_id','accreditation_body','percentage','infra_count']]
```

```python
df_stand_accr_infrastructure_sort_si_2019=df_stand_accr_infrastructure_si_2019.sort_values(['percentage','infra_count'],ascending=False)
```
